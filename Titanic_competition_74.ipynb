```python
import pandas as pd
import numpy as np
```

```python
train_data = pd.read_csv("train.csv")
train_data.head()
test_data2 = pd.read_csv("test.csv")
test_data2.head()
```

```python
test_data = test_data2.copy()
```

```python

X = train_data.drop("Survived", axis=1)
y = train_data["Survived"]
```

```python
X.head(10)
```

```python
# prompt: correlation of columns of the dataframe X, ignore categorical columns

numeric_cols = X.select_dtypes(include=np.number).columns
correlation_matrix = X[numeric_cols].corr()
print(correlation_matrix)
```

```python
X.info()
```

```python
test_data.info()
```

```python
X.drop(["Cabin","PassengerId","Name", "Ticket"], axis=1, inplace=True)
test_data.drop(["Cabin","PassengerId","Name", "Ticket"], axis=1, inplace=True)
```

```python
# prompt: X dataframe has age column with missing values, fill them with the mean
age_mean = X['Age'].mean()
X['Age'].fillna(age_mean, inplace=True)
test_data.fillna(age_mean, inplace=True)
```

```python
# prompt: remove rows with null values in X dataframe from abc column
null_row_indices = X[X['Embarked'].isnull()].index

X.dropna(subset=['Embarked'], inplace=True)
y.drop(null_row_indices, inplace=True)
```

```python
y.info()
```

```python
fare_mean = X['Fare'].mean()
test_data.fillna(fare_mean, inplace=True)
```

```python
X.info()
test_data.info()
```

```python
X['iKnow']= X["SibSp"] + X["Parch"]
test_data['iKnow']= test_data["SibSp"] + test_data["Parch"]
X.drop(["SibSp","Parch"], axis=1, inplace=True)
test_data.drop(["SibSp","Parch"], axis=1, inplace=True)
```

```python
X.info()
test_data.info()
```

```python
categorical_colms = ["Pclass","Sex", "Embarked"]
X = pd.get_dummies(X, columns=categorical_colms)
test_data = pd.get_dummies(test_data, columns=categorical_colms)
```

```python
X.info()
test_data.info()
```

```python
from xgboost import XGBClassifier
```

```python
xgb_model = XGBClassifier(use_label_encoder=False, eval_metric='logloss')
xgb_model.fit(X, y)
```

```python
result=xgb_model.predict(test_data)
```

```python
type(result)
```

```python
# prompt: make a dataframe using 2 column

# Create a DataFrame with two columns
data = {'PassengerId': test_data2["PassengerId"], 'Survived': result}
df = pd.DataFrame(data)

# Print the DataFrame
df.head()
```

```python
df.to_csv("submission.csv", index=False)
```
